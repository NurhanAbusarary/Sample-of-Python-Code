## Importing Packages

```python
import sys
## The utils.py directory
sys.path.insert(0, '/utils')

import utils

import csv
import numpy as np
import pandas as pd
```

## Specify the Directory

```python
## The Powergrading dataset tsv files directory
files_path = 'dataset'
```

## Read the Questions and Model Aswers File

```python
tsv_file = open(f'{files_path}/questions_answer_key.tsv')
read_tsv = csv.reader(tsv_file, delimiter = '\t')

questions = []

for row in read_tsv:
    questions.append(row)
```

```python
df = pd.DataFrame(questions, columns = ['qustionNumber', 'question', 'referenceAnswers', 'Answer_2', 'Answer_3', 'Answer_4',
                                        'Answer_5', 'Answer_6', 'Answer_7', 'Answer_8', 'Answer_9',
                                        'Answer_10'])
```

```python
new_header = df.iloc[0] 

df = df[1:]
```

```python
df.head()
```

```python
df['referenceAnswers'] = df[df.columns[2:]].apply(lambda x: ','.join(x.astype(str)), axis=1)
```

```python
df.head()
```

```python
df.drop(df.iloc[:, 3:], inplace = True, axis = 1) 
```

```python
df.head()
```

```python
df['referenceAnswers'] = df['referenceAnswers'].str.replace(r'None,', '')
```

```python
df.head()
```

```python
df['referenceAnswers'] = df['referenceAnswers'].str.replace(r',None', '')
```

```python
df.head()
```

## Read the Students Aswers File

```python
tsv_file = open(f'{files_path}/studentanswers_grades_698.tsv')
read_tsv = csv.reader(tsv_file, delimiter = '\t')

train_data = []

for row in read_tsv:
    train_data.append(row)
```

```python
train_df = pd.DataFrame(train_data)
```

```python
new_header = train_df.iloc[0] 

train_df = train_df[1:]

train_df.columns = new_header
```

```python
train_df.head()
```

## Merging all Datasets

```python
train_df = train_df.drop(['student'], axis=1)
```

```python
train_df.head()
```

```python
train_df.G1 = train_df.G1.astype(int)
train_df.G2 = train_df.G2.astype(int)
train_df.G3 = train_df.G3.astype(int)
```

```python
col = train_df.loc[: , 'G1':'G3']
```

```python
train_df['avg'] = col.mean(axis = 1)
```

```python
train_df.head(50)
```

```python
train_df['avg'].unique()
```

```python
train_df['avg'] = train_df['avg'].round(decimals=0)
```

```python
train_df.head()
```

```python
train_df['avg'].unique()
```

```python
train_df.avg = train_df.avg.astype(int)
```

```python
train_df.head()
```

```python
train_df = train_df.set_index('Q#')
```

```python
train_df_final = df.join(train_df, on = 'qustionNumber')
```

```python
train_df_final.rename(columns = {'qustionNumber':'id'}, inplace = True) 
```

```python
train_df_final.answer = train_df_final.answer.astype(str)
```

```python
train_df_final.shape
```

```python
train_df_final.tail()
```

```python
def f(train_df_final):
    if train_df_final['avg'] == 1:
        return 1
    else: 
        return 0
        
train_df_final['correct'] = train_df_final.apply(f, axis=1)
```

```python
train_df_final = train_df_final.drop(['G1', 'G2', 'G3', 'avg'], axis=1)
```

```python
train_df_final = train_df_final.astype(str)
```

```python
train_df_final.head()
```

```python
train_df_final.to_csv('data/Powergrading/UNcleaned_powergrading.csv', index = False)
```
